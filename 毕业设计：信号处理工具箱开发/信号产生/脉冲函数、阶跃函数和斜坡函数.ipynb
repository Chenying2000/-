## 脉冲函数、阶跃函数和斜坡函数

```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
t = np.linspace(-1,1,200)
#Generate a time vector in steps of 10 ms
#Define step signal
def step(t):
     r=np.where(t>0.0,1.0,0.0)
     return r
#Define pulse signal
print(t)
def pulse(t):
     r=np.where(t==0.0,1.0,0.0)
     return r
#Define unit ramp function
def ramp(t):
     r=t*np.where(t>0.0,1.0,0.0)
     return r
#Define unit parabola function
def quad(t):
     r = t **2 * np.where(t > 0.0, 1.0, 0.0)
     return r
plt.plot(t,step(t))
plt.plot(t,pulse(t))
plt.plot(t,ramp(t))
plt.plot(t,quad(t))

```
