## 脉冲函数、阶跃函数和斜坡函数

```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
t = np.linspace(-1,1,200)
#Generate a time vector in steps of 10 ms
#Define step signal
def step(t):
     r=np.where(t>0.0,1.0,0.0)
     return r
#Define pulse signal
print(t)
def pulse(t):
     r=np.where(t==0.0,1.0,0.0)
     return r
#Define unit ramp function
def ramp(t):
     r=t*np.where(t>0.0,1.0,0.0)
     return r
#Define unit parabola function
def quad(t):
     r = t **2 * np.where(t > 0.0, 1.0, 0.0)
     return r
```

```python
#Draw the original signal and the restored time-domain waveform of each sampled signal
fig,ax = plt.subplots(constrained_layout=True)
ax.plot(t,step(t),label='step')
ax.plot(t,pulse(t),label='pulse')
ax.plot(t,ramp(t),label='ramp')
ax.plot(t,quad(t),label='quad')
ax.set_title('常见信号生成')
ax.set_xlabel('t/s')
ax.legend(fontsize=12)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
fig.savefig('./常见信号生成.png',dpi=500)
```
