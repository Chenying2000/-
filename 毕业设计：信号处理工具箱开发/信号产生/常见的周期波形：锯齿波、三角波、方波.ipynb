## 常见的周期波形：锯齿波、三角波、方波

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

### 锯齿波

以 10 kHz 的采样率生成 1.5 秒的 50 Hz 锯齿波。对一个方波进行重复计算。

```python
def swatooth_wave(start, end, period, step):

    #start: 波形的起始点
    #end:   波形的结束点
    #period:波形的周期（一个周期内的点数）
    #step:  周期内的步长

    xout = []
    yout = []
    for i in range(start, end, period):
        x = np.arange(i, i + period, step)
        y = np.where(x >= i, x - i , i + period - x)
        xout = np.append(xout, x)
        yout = np.append(yout, y)
    return xout, yout
```

输入已知的波形信息

```python
fs = 10000 #采样频率
f = 50 #信号频率
N = fs/f # 一个周期内的点数
sum_N = 10*N #本案例中只选取了前0.2s的波形
N = np.array(N, dtype=np.int)
sum_N = np.array(sum_N, dtype=np.int)

a,b = swatooth_wave(0,sum_N,N,1)
X = a/10000 #将点数值改为时间和函数值
Y = b/100-1
```

```python
#绘图
fig,ax = plt.subplots(constrained_layout=True)
ax.plot(X,Y)
ax.set_title('周期性锯齿波')
ax.set_xlabel('t/s')
ax.legend(fontsize=12)
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
fig.savefig('./周期性锯齿波.png',dpi=500)
```

### 三角波

```python
t = np.linspace(0,30,1000)
Y = signal.sawtooth(t, 0.5)
```

```python
fig,ax = plt.subplots(constrained_layout=True)
ax.plot(t/150,Y)
ax.set_title('周期性三角波')
ax.set_xlabel('t/s')
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
fig.savefig('./周期性三角波.png',dpi=500)
```

### 方波

```python
def square_wave(start, end, period, step):
    #start: 波形的起始点
    #end:   波形的结束点
    #period:波形的周期（一个周期内的点数）
    #step: 周期内的步长
    #PWM: 调节占空比

    xout = []
    yout = []
    for i in range(start, end, period):
        x = np.arange(i, i + period, step)
        y = np.where(x >= i + period / 2, 1, 0)
        xout = np.append(xout, x)
        yout = np.append(yout, y)
    return xout, yout
```

输入已知的波形信息

```python
fs = 10000 #采样频率
f = 50 #信号频率
N = fs/f # 一个周期内的点数
sum_N = 10*N #本案例只选取了前0.2s的波形
N = np.array(N, dtype=np.int)
sum_N = np.array(sum_N, dtype=np.int)

a,b = square_wave(0,sum_N,N,1)
X = a/10000 
Y = b
```

```python
#绘图
fig,ax = plt.subplots(constrained_layout=True)
ax.plot(X,Y)
ax.set_title('占空比可调的周期性矩形波')
ax.set_xlabel('t/s')
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
fig.savefig('./占空比可调的周期性矩形波.png',dpi=500)
```
