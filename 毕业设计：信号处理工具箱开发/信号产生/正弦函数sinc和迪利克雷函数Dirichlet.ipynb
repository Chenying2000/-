## 正弦函数sinc和迪利克雷函数Dirichlet

### 正弦函数sinc

sinc 函数计算输入向量或矩阵 x 的数学正弦函数。作为时间或空间的函数，sinc 函数是以零为中心、宽度为 2π 并具有单位高度的频率的矩形脉冲的傅里叶逆变换：

$$sinc x =\frac{1}{2\pi}\int_{-\pi}^{\pi}e^{jwx}dw=\begin{cases}
\frac{sin\pi x}{\pi x},&x\neq0,\\1,&x=0.
\end{cases}$$

要绘制值范围从 −5 到 5 的线性间距向量的 sinc 函数，请使用以下命令：

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
x = np.linspace(-5,5,1000)
y = special.sinc(x)
```

```python
#绘图
fig,ax = plt.subplots(constrained_layout=True)
ax.plot(x,y)
ax.set_title('正弦函数sinc（x）')
ax.set_xlabel('t/s')
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False 
fig.savefig('./正弦函数sinc（x）.png',dpi=500)
```

### 迪利克雷函数Dlrichlet

diric 函数计算输入向量或矩阵 x 的 Dirichlet 函数，有时称为周期性正弦函数或混叠正弦函数。Dirichlet 函数由下式定义：

$$D(x) =\begin{cases}
\frac{Nx/2}{Nsin(x/2)},&x\neq 2\pi k,\\(-1)^{k(N-1)},&x=2 \pi k.
\end{cases}k=0,\pm 1,\pm 2,\pm3,...$$

其中 N 是用户指定的正整数。如果 N 为奇数，则 Dirichlet 函数的周期为2π；如果 N 为偶数，则其周期为 4π。此函数的幅值是 1/N 乘以包含 N 个点的矩形窗的离散时间傅里叶变换的幅值。

- 目的：要为 N=7 和 N=8 绘制 0 和 4π 之间的 Dirichlet 函数

```python

import numpy as np
from scipy import special
import matplotlib.pyplot as plt
x = np.linspace(-4*np.pi, 4*np.pi, num=201)
plt.figure(figsize=(8, 8));
for idx, n in enumerate([7,8]):
    plt.subplot(2, 1, idx+1)
    plt.plot(x/np.pi, special.diric(x, n))
    plt.title('diric, n={}'.format(n),horizontalalignment='right',x=1.0)
    plt.xlabel('x/pi')

```
